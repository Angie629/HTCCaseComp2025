```python
import pandas as pd
import re
```

```python
df = pd.read_csv('/Geoscout Facility Data Dump.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
# clean a column named coownder that contains strings with company names and percentages
def clean_coowner(coowner_string):
    cleaned_data = []
    if isinstance(coowner_string, str):
        entries = coowner_string.split(';')
        for entry in entries:
            match = re.match(r'\s*([\d.]+)%\s*-\s*(.+)', entry)
            if match:
                percentage = float(match.group(1))
                company_name = match.group(2).strip()
                cleaned_data.append((percentage, company_name))
    return cleaned_data

df['Co-Owner'] = df['Co-Owner'].apply(clean_coowner)
df.head()
```

```python
# Remove columns with all NaN values
df.dropna(axis=1, how='all', inplace=True)
df.head()
```

```python
# clean latitude and longitude
def clean_latitude(lat_string):
    if isinstance(lat_string, str):
        match = re.match(r'([\d.]+)\s*deg\s*N', lat_string)
        if match:
            return float(match.group(1))
    return lat_string

def clean_longitude(lon_string):
    if isinstance(lon_string, str):
        match = re.match(r'([\d.]+)\s*deg\s*W', lon_string)
        if match:
            return -float(match.group(1))
    return lon_string

df['Latitude'] = df['Latitude'].apply(clean_latitude)
df['Longitude'] = df['Longitude'].apply(clean_longitude)
df.head()
```

```python
# Assuming 'DATE' column exists and is in 'yyyy/MM' format
def format_date(date_str):
    try:
        year, month = map(int, date_str.split('/'))
        return f"01/{month:02}/{year}"  # Set day to 01
    except:
        return date_str  # Return original string if format is invalid

df['Date'] = df['Date'].apply(format_date)
df.head()
```
